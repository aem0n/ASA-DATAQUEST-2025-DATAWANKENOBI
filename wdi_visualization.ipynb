```python
country_info = pd.read_csv("WDI_CSV_2025_07_02/WDIcountry.csv")

merged_income = pd.merge(
    wdi,
    country_info[["Country Code", "Income Group"]],
    on="Country Code",
    how="left"
)

merged_income = merged_income.dropna(subset=["Income Group"])

merged_income
```

```python
import statsmodels.formula.api as smf

# 1ï¸âƒ£ Veri hazÄ±rlÄ±ÄŸÄ±
df = merged_income.dropna(
    subset=[
        "life_expectancy", "gdp_per_capita", "poverty_rate",
        "unemployment_rate", "maternal_mortality",
        "under5_mortality", "birth_rate", "Income Group"
    ]
).copy()

# 0 veya negatif GDP deÄŸerlerini Ã§Ä±kar
df = df[df["gdp_per_capita"] > 0]

# log(GDP per capita) hesapla
df["log_gdp_per_capita"] = np.log(df["gdp_per_capita"])

# Kolon isimlerini uyumlu hale getirelim
df = df.rename(columns={"Income Group": "Income_Group"})

# 2ï¸âƒ£ FormÃ¼l: log(GDP) * gelir grubu + kontrol deÄŸiÅŸkenleri
formula = """
life_expectancy ~ log_gdp_per_capita * C(Income_Group, Treatment('Low income'))
                 + poverty_rate
                 + unemployment_rate
                 + maternal_mortality
                 + under5_mortality
                 + birth_rate
"""

# 3ï¸âƒ£ OLS modeli (robust SE)
model = smf.ols(formula, data=df).fit(cov_type="HC3")
print(model.summary())

# 4ï¸âƒ£ Her gelir grubu iÃ§in log(GDP) eÄŸimini hesapla
coefs = model.params
levels = ["Low income", "Lower middle income", "Upper middle income", "High income"]

def slope_for(level):
    base = coefs["log_gdp_per_capita"]
    key = f"log_gdp_per_capita:C(Income_Group, Treatment('Low income'))[T.{level}]"
    return base + (coefs[key] if key in coefs.index else 0.0)

def p_for(level):
    # Wald testi ile o grubun eÄŸiminin sÄ±fÄ±rdan farkÄ±nÄ± test et
    if level == "Low income":
        L = np.zeros(len(coefs))
        L[list(coefs.index).index("log_gdp_per_capita")] = 1.0
    else:
        name = f"log_gdp_per_capita:C(Income_Group, Treatment('Low income'))[T.{level}]"
        L = np.zeros(len(coefs))
        L[list(coefs.index).index("log_gdp_per_capita")] = 1.0
        L[list(coefs.index).index(name)] = 1.0
    w = model.wald_test(L)
    return float(w.pvalue)

print("\nEtkilerin gelir grubuna gÃ¶re eÄŸimleri (log_gdp_per_capita):")
for lvl in levels:
    print(f"- {lvl:22s} slope = {slope_for(lvl):.3f}   p={p_for(lvl):.3g}")
```

```python
sns.set(style="whitegrid", font_scale=1.1)

plt.figure(figsize=(10, 7))

sns.scatterplot(
    data=df,
    x="log_gdp_per_capita",
    y="life_expectancy",
    hue="Income_Group",
    palette="Set1",
    alpha=0.6,
    s=70,
    edgecolor="white"
)

for i, group in enumerate(["Low income", "Lower middle income", "Upper middle income", "High income"]):
    if group in df["Income_Group"].unique():
        sns.regplot(
            data=df[df["Income_Group"] == group],
            x="log_gdp_per_capita",
            y="life_expectancy",
            scatter=False,
            color=sns.color_palette("Set1")[i],
            label=f"{group} â€” Trend"
        )

plt.title("Relationship Between GDP per Capita and Life Expectancy\n(by Income Group)", fontsize=14, weight="bold")
plt.xlabel("log(GDP per Capita, USD)", fontsize=12)
plt.ylabel("Life Expectancy at Birth (years)", fontsize=12)

plt.legend(title="Income Group", fontsize=10, title_fontsize=11, loc="lower right")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
country_name = "United States"

df_country = merged_income[merged_income["Country Name"] == country_name].copy()

if df_country.empty:
    raise ValueError(f"No data found for country: {country_name}")

income_group = df_country["Income Group"].iloc[0]

sns.set(style="whitegrid", font_scale=1.2)

fig, ax1 = plt.subplots(figsize=(10, 6))

sns.lineplot(
    data=df_country,
    x="Year",
    y="gdp_per_capita",
    color="#1f77b4",
    linewidth=2.5,
    marker="o",
    ax=ax1
)
ax1.set_xlabel("Year", fontsize=12)
ax1.set_ylabel("GDP per Capita (USD)", color="#1f77b4", fontsize=12)
ax1.tick_params(axis="y", labelcolor="#1f77b4")

ax2 = ax1.twinx()
sns.lineplot(
    data=df_country,
    x="Year",
    y="life_expectancy",
    color="#d62728",
    linewidth=2.5,
    marker="o",
    ax=ax2
)
ax2.set_ylabel("Life Expectancy (years)", color="#d62728", fontsize=12)
ax2.tick_params(axis="y", labelcolor="#d62728")

plt.title(f"{country_name}: GDP per Capita and Life Expectancy Over Time",
          fontsize=15, weight="bold", pad=15)
plt.suptitle(f"Income Group: {income_group}",
             fontsize=11, color="gray", y=0.93)

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], color="#1f77b4", lw=2.5, marker="o", label="GDP per Capita (USD)"),
    Line2D([0], [0], color="#d62728", lw=2.5, marker="o", label="Life Expectancy (years)")
]
ax1.legend(handles=legend_elements, loc="upper left", frameon=True)

plt.grid(alpha=0.3)
fig.tight_layout()
plt.show()
```
